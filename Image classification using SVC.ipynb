```python
# Preprocessing data

import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
```

```python
target = [] # empty list to store single digit values for output 
images = [] # to store resized images
flattened_data = [] # to store flat image(1-D data)

DATA_PATH = "/content/images"
CATEGORIES = ["football", "monkey", "sunflower"]

for category in CATEGORIES:
  category_label = CATEGORIES.index(category)
  ITEM_PATH = os.path.join(DATA_PATH, category)
  for img in os.listdir(ITEM_PATH):
    IMAGE_PATH = os.path.join(ITEM_PATH, img )
    image_array = imread(IMAGE_PATH)
    # print(image_array.shape)
    # plt.imshow(image_array)
    resized_image = resize(image_array, (150, 150, 3))
    flat_image = resized_image.flatten()
    flattened_data.append(flat_image)
    images.append(resized_image)
    target.append(category_label)

target = np.array(target)
images = np.array(images)
flattened_data = np.array(flattened_data)
```

```python
# Splitting data into train, test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(flattened_data, target, test_size=0.2, random_state=42)
```

```python
# Using GridSearchCV to find best parameters
from sklearn.model_selection import GridSearchCV
from sklearn import svm

grid_parameter = [
                  {
                      'C': [1, 10, 100, 1000], 'kernel': ['linear']
                   },
                  {
                      'C': [1, 10, 100, 1000], 'kernel': ['rbf'], 'gamma': [0.1, 0.001, 0.0001]
                  }
]
# Using SVC classifier
svc = svm.SVC(probability=True)
clf = GridSearchCV(svc, grid_parameter)
clf.fit(x_train, y_train)
```

```python
# Predicting test data
y_pred = clf.predict(x_test)
y_pred
```

```python
# Actual output of test data
y_test
```

```python
# Checking accuracy score
from sklearn.metrics import accuracy_score, confusion_matrix 

accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
# creating app.py using %%writefile app.py
%%writefile app.py
import streamlit as st
import numpy as np
from skimage.io import imread
from skimage.transform import resize
import matplotlib.pyplot as plt
import pickle
from PIL import Image
st.title("Image Classifier using Machine Learning")
st.text('Upload the Image')

model = pickle.load(open('image_model.p', 'rb'))
uploaded_file = st.file_uploader("Choose an image...", type='jpg')
if uploaded_file is not None:
  img = Image.open(uploaded_file)
  st.image(img, caption='Uploaded Image')

  if st.button('PREDICT'):
    CATEGORIES = ["football", "monkey", "sunflower"]
    st.write('Result....')
    flat_image = []
    image = np.array(img)
    image_resized = resize(image, (150, 150, 3))
    flat_image.append(image_resized.flatten())
    flat_image = np.array(flat_image)
    plt.imshow(image_resized)
    output = model.predict(flat_image)
    output = CATEGORIES[output[0]]
    st.title(f'Predicted Output: {output}')
    prob = model.predict_proba(flat_image)
    for index, item in enumerate(CATEGORIES):
      st.write(f"{item}: {prob[0][index]*100}")
```
